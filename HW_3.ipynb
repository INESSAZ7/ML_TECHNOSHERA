## Отображение школы

```python
center = {'lat': 54.875414, 'lon': 37.2052787}
```

```python
map_folium = folium.Map(location=(center['lat'], center['lon']),
                        tiles='Stamen Toner',
                        zoom_start=11)
```

```python
norm = mpl.colors.Normalize(vmin=0, vmax=50000)
cmap = cm.plasma
colormap= cm.ScalarMappable(norm=norm, cmap=cmap)
for ind, row in df[df['Unnamed: 0'] > 42580].iterrows():
    folium.CircleMarker([row['lat'], row['lon']], 
                        popup='{}'.format('<br>'.join(map(str, [row['name']]))),
                        fill=True).add_to(map_folium)

map_folium
```
